# Data Quality Evaluation
## Data Description

* The dataset contains the national average response to questions about quality of life for 165 countries from 2005 to 2022.
* Only about 50 countries have data for all 17 years, but the majority of countries have 10+ years of observations with 2199 total rows.
* 5 columns are averages of binary responses:
    * **Social Support**: If you were in trouble, do you have relatives or friends you can count on to help you whenever you need them, or not?
    * **Freedom to make life choices**: Are you satisfied or dissatisfied with your freedom to choose what you do with your life?
    * **Generosity**: Have you donated money to a charity in the past month?
    * **Perceptions of corruption**: "Is corruption widespread throughout the government and businesses or not
    * **Positive affect**: Did you smile or laugh a lot yesterday?
    * **Negative affect**: Did you experience the following feelings during A LOT OF THE DAY yesterday? How about Sadness, Worry, Anger?
* **Healthy life expectancy at birth** represents in years the life expectancy in given country
* **Life Ladder** is the average response to question Please imagine a ladder, with steps numbered from 0 at the bottom to 10 at the top. The top of the ladder represents the best possible life for you and the bottom of the ladder represents the worst possible life for you. On which step of the ladder would you say you personally feel you stand at this time?
* **Log GDP per capita** is log transformed GDP of the given country in the given year.
* Not many missing values in the dataset. **Generosity** has 73 missing rows and **Perceptions of corruption** has 116, but no other column has more than about 50 missing rows.
* 2005 and 2006 have significantly less observations than the other years
* 2013+ consistenly has 130+ observations.

# Question 2

```python
def calculate_happiness(df):
    numeric_columns = df.select_dtypes(include=['number'])
    non_numeric_columns = df.select_dtypes(exclude=['number'])

    # Flip the values for negative affect in order to keep it consistent with other columns
    numeric_columns["Negative affect"] = 1 - numeric_columns["Negative affect"]
    
    normalized_numeric = (numeric_columns - numeric_columns.mean()) / numeric_columns.std()
    normalized_numeric.drop(['year'], axis=1, inplace=True)
    normalized_numeric = normalized_numeric.mean(axis=1, skipna=True)
    final_df = pd.concat([normalized_numeric, non_numeric_columns], axis=1)
    return final_df.sort_values(by=0, ascending=False)
```

```python
df_2022 = df[df["year"] == 2022]
calculate_happiness(df_2022)
```

```python
df_2013 = df[df["year"] == 2013]
calculate_happiness(df_2013)
```

```python
df_all_time = df.groupby("Country name").mean()
calculate_happiness(df_all_time)
```

To create a metric for happiest country, I calculated the z score for each numeric column, took the average, and sorted in a descending order. By averaging the z score, I can see which countries consistently stand out both positively and negatively. To do this I had to flip the value for positive affect in order to keep the higher the number the better the meaning consistent.

# Question 3

```python
# get first year country appeared
min_index = df.groupby('Country name')['year'].idxmin()
first_df = df.loc[min_index]

# get last year country appeared
max_index = df.groupby('Country name')['year'].idxmax()
last_df = df.loc[max_index]

happiest_first = calculate_happiness(first_df)
happiest_last = calculate_happiness(last_df)

merged_df = pd.merge(happiest_first, happiest_last, on="Country name", how="inner")

merged_df['Difference'] = merged_df['0_y'] - merged_df['0_x']
result_df = merged_df[['Country name', 'Difference']]
result_df = result_df.sort_values(by='Difference', ascending=False)
result_df
```

To answer question 3, I created two data frames, one for the first year each country appeared, and one for the last year. I then calculated the happiness for each df in the same manner as before. I then subtracted the first happiness value from the last to get the biggest difference. I found Serbia had the biggest positive difference, and Afghanistan had the biggest negative difference.
